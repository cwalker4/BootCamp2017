## Temperatures

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from datetime import datetime

chi = pd.read_csv("chicago.csv", index_col='DATE', na_values=-9999)
chi.index = pd.to_datetime(chi.index, format='%Y%m%d')
chi.index = chi.index.dayofyear
chi = chi.drop(["STATION","ELEVATION","LATITUDE","LONGITUDE"], axis=1)


ind = pd.read_csv("indianapolis.csv", index_col="DATE", na_values=-9999)
ind.index = pd.to_datetime(ind.index, format='%Y%m%d')
ind.index = ind.index.dayofyear
ind = ind.drop(["STATION","STATION_NAME"], axis=1)


mia = pd.read_csv("miami.csv", index_col="DATE", na_values=-9999)
mia.index = pd.to_datetime(mia.index, format='%Y%m%d')
mia.index = mia.index.dayofyear
mia = mia.drop(["STATION","ELEVATION", "LATITUDE", "LONGITUDE"], axis=1)

pit = pd.read_csv("pittsburgh.csv", index_col="DATE", na_values=-9999)
pit.index = pd.to_datetime(pit.index, format='%Y%m%d')
pit.index = pit.index.dayofyear
pit = pit.drop(["STATION","STATION_NAME"], axis=1)

was = pd.read_csv("washington.csv", index_col="DATE", na_values=-9999)
was.index = pd.to_datetime(was.index, format='%Y%m%d')
was.index = was.index.dayofyear
was = was.drop(["STATION","STATION_NAME"], axis=1)

x = np.arange(1,367)

plt.figure(figsize=(20,9))
plt.plot(was, '.', color='black', ms=2)
plt.plot(pit, '.', color='black', ms=2)
plt.plot(mia, '.', color='black', ms=2)
plt.plot(ind, '.', color='black', ms=2)
plt.plot(chi, '.', color='#800000', ms=2)
plt.ylabel("Temperature (F)")
x = [80,172, 266, 356]
labels=['Fall', 'Spring', 'Summer', 'Winter']
plt.xticks(x, labels)

plt.show()
```
