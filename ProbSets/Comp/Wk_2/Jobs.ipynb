```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv("payems.csv", index_col='date', skiprows=[0,1,2,3,4])
df.index = pd.to_datetime(df.index, format='%m/%d/%y')

r1 = df['2028-08-01':'2036-08-01']
r2 = df['2036-05-01':'2044-05-01']
r3 = df['2044-02-01':'2052-02-01']
r4 = df['2047-11-01':'2055-11-01']
r5 = df['2052-07-01':'2060-07-01']
r6 = df['2056-08-01':'2064-08-01']
r7 = df['2059-04-01':'2067-04-01']
r8 = df['1969-01-01':'1977-01-01'] #transition period 
r9 = df['1972-11-01':'1980-11-01']
r10 = df['1979-01-01':'1987-01-01']
r11 = df['1980-07-01':'1988-07-01']
r12 = df['1989-07-01':'1997-07-01']
r13 = df['2000-03-01':'2008-03-01']
r14 = df['2006-12-01':'2014-12-01']


r1 = r1 / r1.iloc[0]
r2 = r2 / r2.iloc[1] 
r3 = r3 / r3.loc['2045-02-01']
r4 = r4 / r4.loc['2048-11-01']
r5 = r5 / r5.loc['2053-07-01']
r6 = r6 / r6.loc['2057-08-01']
r7 = r7 / r7.loc['2060-04-01']
r8 = r8 / r8.loc['1969-12-01']
r9 = r9 / r9.loc['1973-11-01']
r10 = r10 / r10.loc['1980-01-01']
r11 = r11 / r11.loc['1981-07-01']
r12 = r12 / r12.loc['1990-07-01']
r13 = r13 / r13.loc['2001-03-01']
r14 = r14 / r14.loc['2007-12-01']

#for ar in (r1,r2,r3,r4,r5,r6,r7,r8,r9,r10,r11,r12,r13,r14):
#    jobs_peak = ar.max()
#    ar = ar / jobs_peak

x = np.linspace(0,8,97)
x_r2 = np.linspace(0,8,68)
x_r1 = np.linspace(1,8,8)

plt.figure(figsize=(10,5))
plt.plot(x_r1, r1, color='black', lw=3, label='August 1929')
plt.plot(x_r2, r2, label='May 1937')
plt.plot(x, r3, label='February 1945')
plt.plot(x, r4, label='November 1948')
plt.plot(x, r5, label='July 1953')
plt.plot(x, r6, label='August 1957')
plt.plot(x, r7, label='April 1960')
plt.plot(x, r8, label='December 1969')
plt.plot(x, r9, label='November 1973')
plt.plot(x, r10, label='January 1980')
plt.plot(x, r11, label='July 1981')
plt.plot(x, r12, label='July 1990')
plt.plot(x, r13, label='March 2001')
plt.plot(x, r14, color='red', lw=3, label='December 2007')
x = [0, 1, 2, 3, 4, 5, 6, 7, 8]
labels = ['-1yr', 'peak', '+1yr', '+2yr', '+3yr', '+4yr', '+5yr', '+6yr', '+7yr']
plt.xticks(x, labels)
plt.ylabel('Jobs/peak')
plt.xlabel('Time from peak')
plt.legend(bbox_to_anchor=(1.05,1), loc=2, borderaxespad=0)
plt.axvline(x=1, color='0.5', linestyle='--', lw=1)
plt.axhline(y=1, color='0.5', linestyle='--', lw=1)

plt.show()
```
