```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def rand_array(n):
    a = np.random.randn(n,n)
    mean = np.mean(a, axis=0)
    var = np.var(mean)
    return var

def disp_array():
    args = np.arange(100, 1100, 100)
    var = np.zeros(10)
    for i in range(10):
        var[i] = rand_array(args[i])
    plt.plot(var)
    plt.show()
```

```python
disp_array()
```

```python
def trig_plots():
    x = np.linspace(-2*np.pi, 2*np.pi, 1000)
    sin = np.sin(x)
    cos = np.cos(x)
    atan = np.arctan(x)
    plt.plot(x, sin)
    plt.plot(x, cos)
    plt.plot(x, atan)
    plt.show()
```

```python
trig_plots()
```

```python
def func_plot():
    left = np.linspace(-2, 0.99, 1000)
    right = np.linspace(1.001, 7, 1000)
    f1 = 1 / (left - 1)
    f2 = 1 / (right - 1)
    x = np.arange(-2, 7)
    plt.plot(left, f1, 'm--', lw=2)
    plt.plot(right, f2, 'm--', lw=2)
    plt.ylim(-6,6)
    plt.show()
```

```python
func_plot()
```

Problem 4:

```python
def trig_subplots():
    x = np.linspace(0, 2*np.pi)
    f1 = np.sin(x)
    f2 = np.sin(2*x)
    f3 = 2*np.sin(x)
    f4 = 2*np.sin(2*x)
    
    # plotting sin(x)
    plt.subplot(221)
    plt.plot(x, f1, 'g-')
    plt.axis([0, 2*np.pi, -2, 2])
    plt.title('sin(x)')
    plt.tick_params(labelbottom="off")
    
    # plotting sin(2x)
    plt.subplot(222)
    plt.plot(x, f2, 'r--')
    plt.axis([0, 2*np.pi, -2, 2])
    plt.title('sin(2x)')
    plt.tick_params(labelleft="off")
    plt.tick_params(labelbottom="off")
    
    # plotting 2sin(x)
    plt.subplot(223)
    plt.plot(x, f3, 'b--')
    plt.axis([0, 2*np.pi, -2, 2])
    plt.title('2sin(x)')
    
    # plotting 2sin(2x)
    plt.subplot(224)
    plt.plot(x, f4, 'm:')
    plt.axis([0, 2*np.pi, -2, 2])
    plt.title('2sin(2x)')
    plt.tick_params(labelleft="off")
     
    plt.suptitle('Some sin functions')
    plt.show()
```

```python
trig_subplots()
```

Problem 5:

```python
def fars_plots():
    a = np.load('FARS.npy')
    
    plt.figure(figsize=(10, 15))

    plt.subplot(211)
    longs = a[:,1]
    lats = a[:,2]
    plt.plot(longs, lats, "k,")
    plt.ylabel("Latitude")
    plt.xlabel("Longitude")
    plt.axis('equal')
    
    
    plt.subplot(212)
    plt.xlim(0,24)
    plt.xlabel("Hour of the day")
    hours = a[:,0]
    plt.hist(hours, bins=np.arange(0,25))
    
    plt.tight_layout()
    plt.show()

    
```

```python
fars_plots()
```

Problem 6:

```python
def plot_fun_funcs():
    x = np.linspace(-2*np.pi, 2*np.pi, 1000)
    y = np.linspace(-2*np.pi, 2*np.pi, 1000)
    X, Y = np.meshgrid(x, y)
    Z = (np.sin(X)*np.sin(Y))/(X*Y)
    
    plt.figure(figsize=(12, 5))

    plt.subplot(121)
    plt.pcolormesh(X, Y, Z, cmap="viridis")
    plt.xlim([-2*np.pi, 2*np.pi])
    plt.ylim([-2*np.pi, 2*np.pi])
    plt.colorbar()
    
    plt.subplot(122)
    plt.contour(X, Y, Z, 10, cmap="Spectral")
    plt.xlim([-2*np.pi, 2*np.pi])
    plt.ylim([-2*np.pi, 2*np.pi])
    plt.colorbar()
    
    plt.tight_layout()
    plt.show()
    
    
```

```python
plot_fun_funcs()
```
